## Question 6
Harder - Which Beverage_prep type contains more than the average calories of all drinks? (Hint: to answer this, you’ll have to first figure out what the average calories across all drinks are, and then use that as a filter for the grouped Beverage_prep data).

```python
average_calories = drinks['Calories'].mean()

mean_beverage_prep_calories = (
    drinks
    .groupby("Beverage_prep")["Calories"]
    .mean()
    .reset_index()
    .copy()
)
#print(mean_beverage_prep_calories)
beverage_greater_avg_calories = (
    mean_beverage_prep_calories
  #  .loc[mean_beverage_prep_calories["mean_calories"] > average_calories]
    .loc[mean_beverage_prep_calories["Calories"] > average_calories]
  #  .reset_index(name = "mean_calories")
   .copy()
)
print(beverage_greater_avg_calories)

#print(beverage_greater_avg_calories)
```
